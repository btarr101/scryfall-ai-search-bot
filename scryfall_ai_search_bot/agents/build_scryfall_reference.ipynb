```python
import httpx

http_client = httpx.Client()

search_reference_response = http_client.get("https://scryfall.com/docs/syntax")
search_reference_response.raise_for_status()
search_reference_html = search_reference_response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(search_reference_html)
main_html = soup.find(class_="main")
```

```python
import html2text
from IPython.display import Markdown

markdown = html2text.html2text(str(main_html))

with open("scryfall_search_reference.md", "w+") as f:
	f.write(markdown)

display(Markdown(markdown))
```
